```python
# Hyper parameter

learning_rate = 0.01
epoch = 1000000
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X_train = np.random.rand(100)
```

```python
print("Train data shape :", X_train.shape)
print("Train data type :", X_train.dtype)
```

```python
y_train = 4 * X_train + np.random.rand(100)
```

```python
print("Target data shape :", X_train.shape)
print("Target data type :", X_train.dtype)
```

```python
plt.scatter(X_train, y_train)
```

#### STEP 2. Gradient descent function

```python
Image('/Users/imlim/Downloads/Git/DL/Tensorflow/SLR_gradient.png', width = 250)
```

```python
def gradient_w(y_pred, y, x):
    N = y.shape[0]
    return -2 / N * np.sum(np.dot((y-y_pred),x))
    
def gradient_b(y_pred, y):
    N = y.shape[0]
    return -2 / N * np.sum(y-y_pred)
    
```

#### STEP 3. Algorithm

```python
w0, b = np.array([3], dtype ='float64'), np.array([3], dtype ='float64')
```

```python
def cal_coef(x, y, epoch, learning_rate, w0, b):
    for i in range(epoch):
        y_pred = w0 * x + b
        loss_w = gradient_w(y_pred, y, x)
        loss_b = gradient_b(y_pred, y)
        
        w0 -= learning_rate * loss_w
        b -= learning_rate * loss_b
        
    return w0, b
```

```python
w0, b = cal_coef(X_train, y_train, epoch, learning_rate,w0,b)
print("weight : {:.2f}, bias : {:.2f}".format(float(w0), float(b)))
```

```python
y_pred = w0 * X_train + b

plt.scatter(X_train, y_train)
plt.plot(X_train, y_pred, 'r')
```
